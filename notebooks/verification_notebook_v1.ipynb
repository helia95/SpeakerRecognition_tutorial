```python
import os
import numpy as np

import torch
import torch.nn.functional as F
```

```python
path_embd_test = 'vox_celeb_embeddings/test/'
path_embd_embd =  'vox_celeb_embeddings/enroll/'
```

```python
def load_embeddings(path):
    X = {}
    for file in os.listdir(path):
        th_embd = torch.load(os.path.join(path, file))        
        X[file] = th_embd
        
    return X
```

```python
test_embd = load_embeddings(path_embd_test)
enroll_embd = load_embeddings(path_embd_enroll)
```

```python
users = list(test_embd.keys())
```

```python
users
```

```python
track_matrices = {}
overall_acc = 0
for enroll_user in users:
    print('='*50)
    metrics = {'TP':0, 'FP':0, 'FN': 0, 'TN': 0}
    
    max_score = -10**8
    best_spk = None
    
    for test_user in users:
        reference = enroll_embd[enroll_user]

        X = test_embd[test_user]

        scores = []
        for to_test in X:

            score = F.cosine_similarity(reference, to_test.unsqueeze(0))
            score = score[0].data.cpu().numpy()


        print('\n')
        if score > 0.95:
            pred = 'Accept'
        else:
            pred = 'Reject'

        
        #gt = ''
        if test_user == enroll_user and pred == 'Accept':
            gt = 'True Positive'
            metrics['TP'] += 1
        elif test_user == enroll_user and pred == 'Reject':
            gt = 'False negative'
            metrics['FN'] += 1
        elif not test_user == enroll_user and pred == 'Accept':
            gt = 'False positive'
            metrics['FP'] += 1
        else:
            gt = 'True Negative'
            metrics['TN'] +=1
        
        if score > max_score:
            best_spk = test_user
            max_score = score
            print('Best user updated!')
        
        track_matrices[enroll_user] = metrics
        
        
        message = f'True user: {enroll_user}\nTest user: {test_user}\nScore: {score :.2f}\nResult: {pred}, Metric: {gt}'
        print(message)
    
    print(f'True user: {enroll_user}, Final pred user: {best_spk}')
    if enroll_user == best_spk:
        overall_acc += 1

   
    
```

```python
overall_acc / len(users)
```
