```python
import os
import numpy as np
import librosa
```

```python
base = 'vox_celeb_test_set/original/'
users = os.listdir(base)
```

```python
n_users = len(users)
```

```python
video_to_merge = {}

for user in users:
    videos = os.listdir(os.path.join(base, user))
    count = 0
    duration = 0
    
    n_videos = {}
    for video in videos: 
        n_videos[video] = len(os.listdir(os.path.join(base, user, video)))
        
    video_to_merge[user] = n_videos
```

```python
video_to_extract = {}
for user, stats in video_to_merge.items():
        
    
    enroll_video = max(stats, key=stats.get)
    
    del stats[enroll_video]
    
    test_video = max(stats, key=stats.get)
    

    

    
    video_to_extract[user] = {
        'enroll':os.path.join(base, user, enroll_video),
        'test':os.path.join(base, user, test_video)
    }
```

```python
video_to_extract
```

```python
import shutil
```

```python
out_path ='vox_celeb_test_set/merged/'
```

```python
sr = 16000
for user, videos in video_to_extract.items():
    print(user)
    
    path_usr = os.path.join(out_path, user)
    shutil.rmtree(path_usr, ignore_errors=True)
    os.makedirs(path_usr)
    
    for step, path in videos.items():
        
        files = os.listdir(path)
        
        Y = []
        for file in files:
            
            y, _ = librosa.load(os.path.join(path, file), sr=sr)
            Y.append(y)
            
        Y_ts = np.concatenate(Y)
        
        print(step, librosa.get_duration(Y_ts, sr=sr))
        librosa.output.write_wav(os.path.join(path_usr, f'{step}.wav'), Y_ts, sr, norm=False)
    print('='*50)
```
